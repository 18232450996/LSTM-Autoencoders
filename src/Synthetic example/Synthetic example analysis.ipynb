```python
%matplotlib inline
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

# Analysis

```python
# comparation of errors of the with- ad without-retraining models
with_error_n = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/always_retraining/normal_retraining(batch).csv")
with_error_a = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/always_retraining/abnormal_retraining(batch).csv")
with_all = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/always_retraining/all_retraining(batch).csv")

without_error_n = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/without_retraining/normal_without_retraining(batch).csv")
without_error_a = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/without_retraining/abnormal_without_retraining(batch).csv")
without_all = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/without_retraining/all_without_retraining(batch).csv")

p1_error_n = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/retrain_p1/normal_retraining(batch).csv")
p1_error_a = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/retrain_p1/abnormal_retraining(batch).csv")
p1_all = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/retrain_p1/all_retraining(batch).csv")

with_buffer_n = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/always_retraining_withDataBuffer/normal_retraining(batch).csv")
with_buffer_a = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/always_retraining_withDataBuffer/abnormal_retraining(batch).csv")
with_buffer_all = pd.read_csv("C:/Users/Bin/Desktop/Thesis/Synthetic example (Power demand dataset)/always_retraining_withDataBuffer/all_retraining(batch).csv")
```

```python
plt.figure(figsize=(20,5))
ax = plt.subplot(111)
x = np.array((range(with_error_n.shape[0]+with_error_a.shape[0])))
ax.bar(x-0.3,without_all['max'],width=0.2,color='b',align='center',label='without_retraining')
ax.bar(x-0.1,p1_all['max'],width=0.2,color='y',align='center',label='retrain(when error.max>error.max[0])')
ax.bar(x+0.1,with_all['max'],width=0.2,color='r',align='center',label='always_retraining')
ax.bar(x+0.3,with_buffer_all['max'],width=0.2,color='green',align='center',label='always_retraining_with_all_seen_data')

#pd.Series(np.ones(p1_all['max'].size)*p1_all['max'][1]).plot(label='error.max[0]',color='grey')
ax.autoscale(tight=True)
plt.legend()
plt.xlabel("Batch number")
plt.ylabel("Reconstruction error")
plt.title("Max reconstruction error")

plt.show()
```

```python
plt.figure(figsize=(20,5))
ax = plt.subplot(111,axisbg='lavender')
x = np.array((range(with_error_n.shape[0])))
ax.bar(x-0.3,without_error_n['max'],width=0.2,color='b',align='center',label='without_retraining')
ax.bar(x-0.1,p1_error_n['max'],width=0.2,color='y',align='center',label='retrain(when error.max>error.max[0])')
ax.bar(x+0.1,with_error_n['max'],width=0.2,color='r',align='center',label='always_retraining')
ax.bar(x+0.3,with_buffer_n['max'],width=0.2,color='green',align='center',label='always_retraining_with_all_seen_data')

#pd.Series(np.ones(p1_error_n['max'].size)*p1_error_n['max'][0]).plot(label='error.max[0]',color='grey')
#ax.autoscale(tight=True)
plt.legend()
plt.ylim(0,0.5)
plt.xlabel("Batch number")
plt.ylabel("Reconstruction error")
plt.title("Max reconstruction error of each normal batches")

plt.show()
```
