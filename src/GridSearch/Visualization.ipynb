```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from numpy.random import randn
import pandas as pd
import numpy as np
```

# HTTP

```python
f = open("C:/Users/Bin/Desktop/Thesis/code/src/GridSearch/http_results_.txt","r")
data = []
for _ in range(10):
    tmp = []
    for _ in range(10):
        tmp.append(float(f.readline().strip("[|]\n").split(", ")[-1]))
    data.append(tmp)
    
up, down = float("%.2f" %(np.array(data).ravel().max())), float("%.2f" %(np.array(data).ravel().min()))

fig, ax = plt.subplots()
cax = ax.imshow(data, interpolation='lanczos', cmap=cm.magma)
ax.set_title('Grid search (HTTP)')
ax.set_xticks(range(0,10))
ax.set_xticklabels(range(10,110,10))
ax.set_xlabel("window length")
ax.set_yticks(range(0,10))
ax.set_yticklabels(range(5,55,5))
ax.set_ylabel("Hidden layer size")
cbar = fig.colorbar(cax, ticks=[down,up])
cax.set_clim(down,up)
cbar.ax.set_yticklabels([str(down),str(up)])  
cbar.ax.set_ylabel("REratio")
plt.show()
```
