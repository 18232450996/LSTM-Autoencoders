```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import cm
from numpy.random import randn
import pandas as pd
import numpy as np
```

```python
f = open("C:/Users/Bin/Desktop/Thesis/code/src/GridSearch/http_results.txt","r")
data = []
for _ in range(10):
    tmp = []
    for _ in range(10):
        tmp.append(float(f.readline().strip("[|]\n").split(", ")[-1]))
    data.append(tmp)
    
```

```python
np.array(data).ravel().mean()
```

```python
# Make plot with vertical (default) colorbar
fig, ax = plt.subplots()

cax = ax.imshow(data, interpolation='lanczos', cmap=cm.magma)
ax.set_title('Grid search (HTTP)')
ax.set_xticks(range(0,10))
ax.set_xticklabels(range(10,110,10))
ax.set_xlabel("window length")
ax.set_yticks(range(0,10))
ax.set_yticklabels(range(1,11))
ax.set_ylabel("Hidden layer size")
cbar = fig.colorbar(cax, ticks=[0.5,1,1.5,2,2.5])
cbar.ax.set_yticklabels(['0.5','1.0',"1.5","2.0","2.5"])  
cbar.ax.set_ylabel("Error(normal) / Error(anomaly)")
plt.show()
```
