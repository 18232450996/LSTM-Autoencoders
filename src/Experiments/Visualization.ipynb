```python
time = pd.DataFrame([[32,2],[30,5],[476,36],[929,63]],columns=["adaptive","stationary"],index=["Power","SMTP","HTTP","SMTP+HTTP"])
```

```python
ax = time.plot(kind='bar',figsize=(15,7),fontsize=13)
ax.set_title("Run time comparation ")
for i in ax.patches:
    # get_x pulls left or right; get_height pushes up or down
    ax.text(i.get_x()+.1, i.get_height()+20, \
            str(round((i.get_height()), 2))+"s", fontsize=11, color='dimgrey')
```
