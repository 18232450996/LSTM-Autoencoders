# Decoder side input experiments

# Decoder input is the real value of previous time stamp (first timestep's input is a  zero-tensor) bias =0.1

```python
data = pd.read_csv("C:/Users/Bin/Desktop/Thesis/tmp/EncDecADModel_online_init/power_demand/Try1_2_40_84.csv")
data[:200].plot(figsize=(20,5))

plt.title("Power demand: batch_num=2, #hidden layer units=40, time_steps=84")
```

```python
err =abs( data['original']- data['reconstructed'])
err.plot()
plt.ylim(0,0.8)
plt.xlim(-50,err.shape[0])
plt.title("Reconstruction error")
plt.plot()
```
