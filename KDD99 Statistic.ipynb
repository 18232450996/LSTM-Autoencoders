```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# 10% KDD99

```python
filename = "C:/Users/Bin/Documents/Datasets/KDD99/kddcup.data_10_percent_corrected"
```

```python
normal = 0
anomaly = 0
n = []
a = []
for chunk in pd.read_csv(filename, chunksize=10000):
    n.append(normal)
    a.append(anomaly)
    print(normal,anomaly)
    for index,row in chunk.iterrows():
        if row[-1] == "normal.":
            normal +=1
        else: 
            anomaly += 1
```

```python
import matplotlib.pyplot as plt
pd.Series(n).plot(label="Normal")
pd.Series(a).plot(label="anomaly")
plt.title("Statistic of 10% KDD99 dataset")
plt.xlabel("Index (in 10,000)")
plt.legend()
plt.show()
```

# KDD99

```python
filename = "C:/Users/Bin/Documents/Datasets/KDD99/kddcup.data.corrected"
```

```python
normal = 0
anomaly = 0
n_ = []
a_ = []
for chunk in pd.read_csv(filename, chunksize=10000):
    n_.append(normal)
    a_.append(anomaly)
    #print(normal,anomaly)
    for index,row in chunk.iterrows():
        if row[-1] == "normal.":
            normal +=1
        else: 
            anomaly += 1
```

```python
len(n_)
```

```python
len(a_)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,6))
pd.Series(n_).plot(label="Normal")
pd.Series(a_).plot(label="anomaly")
plt.title("Statistic of KDD99 dataset")
plt.xlabel("Index (in 10,000)")
plt.legend()
plt.show()
```

```python
n_d = np.array([n_[i]-n_[i-1] for i in range(1,len(n_))])
```

```python
for i in range(n_d.size):
    if n_d[i] != 0:
        n_d[i] = 1
```

```python
plt.figure(figsize=(18,6))
plt.scatter(x=range(n_d.size),y=range(0,n_[-1]+a_[-1],(n_[-1]+a_[-1])//n_d.size),s = 2,c='r',label="anomaly")
plt.scatter(x=range(n_d.size),y=n_d*range(0,n_[-1]+a_[-1],(n_[-1]+a_[-1])//n_d.size),s=2,c='b',label="normal")
plt.title("Statistic of KDD99 dataset")
plt.xlabel("Index (in 10,000)")
plt.legend()
plt.show()
```
